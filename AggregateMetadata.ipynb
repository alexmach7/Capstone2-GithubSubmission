```python
set(df23.columns) - set(df24.columns)
set(df24.columns) - set(df23.columns)
```

```python
df24 = df24.drop(columns=["Unnamed: 3", "Unnamed: 4"], errors="ignore")
```

```python
df23["source"] = "dataset_23"
df24["source"] = "dataset_24"
```

```python
rename_map = {
    "Date.grain.collected": "Harvest.date",
    "Replicate": "Replicate_No",
    "Selected.for.MYCOTOXIN.TESTING": "Selected.for.mycotoxin.testing"
}

df23 = df23.rename(columns=rename_map)
df23.columns
```

```python
df23.columns.equals(df24.columns)
```

```python
missing = df23.columns.difference(df24.columns)
print("Missing in df24:", list(missing))

# add them as empty columns
for c in missing:
    df24[c] = pd.NA

# now reorder df24 to match df23
df24 = df24[df23.columns]

# confirm
print(df23.columns.equals(df24.columns))
```

```python
set(df23.columns) == set(df24.columns)
```

Add year label before stacking

```python
df23["source"] = "dataset_23"
df24["source"] = "dataset_24"
```

Order columns

```python
df24 = df24[df23.columns]
```

Combine and Sanity Checks

```python
# Stack (row-bind)
combined_df = pd.concat([df23, df24], ignore_index=True)

print("df1 shape:", df23.shape)
print("df2 shape:", df24.shape)
print("combined shape:", combined_df.shape)
```

```python
master_df = combined_df
master_df.to_csv("master_df.csv", index=False)
```
