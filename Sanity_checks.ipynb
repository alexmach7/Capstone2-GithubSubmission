```python
import pandas as pd
```

```python
df = pd.read_csv('Final2023_2024daily_avg.csv')
```

```python
df.describe()
```

NDVI

```python
import matplotlib.pyplot as plt

# Pick a site that definitely exists
site_name = df['Location'].unique()[0]
site_df = df[df['Location'] == site_name].sort_values('date')

# Confirm dates exist
print(site_df[['date', 'NDVI']].head())

# Plot
plt.figure(figsize=(12, 5))
plt.plot(site_df['date'], site_df['NDVI'], marker='o', linestyle='-', color='green', alpha=0.7)
plt.title(f"NDVI Over Time - {site_name}")
plt.xlabel("Date")
plt.ylabel("NDVI")
plt.ylim(0, 1)
plt.grid(True)
plt.tight_layout()
plt.xticks(rotation=45)  # Rotate x-axis labels
plt.show()
```

```python
locations_to_plot = df['Location'].unique()

plt.figure(figsize=(14, 6))
for loc in locations_to_plot:
    subset = df[df['Location'] == loc].sort_values('date')
    plt.plot(subset['date'], subset['NDVI'], label=loc, alpha=0.8)

plt.title("NDVI Over Time for Selected Locations")
plt.xlabel("Date")
plt.ylabel("NDVI")
plt.ylim(0, 1)  # Adjust if needed
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
locations_to_plot = df['Location'].unique()[:3]  # First 3 locations

plt.figure(figsize=(14, 6))
for loc in locations_to_plot:
    subset = df[df['Location'] == loc].sort_values('date')
    plt.plot(subset['date'], subset['NDVI'], label=loc, alpha=0.8)

plt.title("NDVI Over Time for Selected Locations")
plt.xlabel("Date")
plt.ylabel("NDVI")
plt.ylim(0, 1)  # Adjust if needed
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

RH

```python
# Replace with any Location you want to plot
site_name = df['Location'].unique()[0]

# Filter data for that location
site_df = df[df['Location'] == site_name].sort_values('date')

# Plot
plt.figure(figsize=(12, 5))
plt.plot(site_df['date'], site_df['RH'], marker='o', linestyle='-', alpha=0.7)
plt.title(f"Relative Humidity Over Time - {site_name}")
plt.xlabel("Date")
plt.ylabel("Relative Humidity")
plt.grid(True)
plt.tight_layout()
plt.show()
```

Temperature

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 5))
for loc in df['Location'].unique():
    subset = df[df['Location'] == loc]
    plt.plot(subset['date'], subset['temperature'], label=loc)

plt.xlabel('Date')
plt.ylabel('Temperature')
plt.title('Temperature Over Time by Location')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
